```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
data = pd.read_csv("apple_products.csv")
```

```python
 data.describe()
```

look at whether this dataset contains any null values

```python
data.isnull().sum()
```

storing all the data about the top 10 highest-rated iPhones


```python

highest_rated = data.sort_values(by=["Star Rating"], ascending=False)
```

```python
highest_rated['Product Name'].head(10)
```

```python
data.sort_values(by=["Number Of Reviews"], ascending= False)
```

 Now let’s have a look at the number of ratings of the highest-rated iPhones

```python
highest_rated = highest_rated.head(10)

iphones = highest_rated["Product Name"].value_counts()
label = iphones.index
counts = highest_rated["Number Of Ratings"]

plt.figure(figsize=(6, 3))   
plt.bar(label, counts)
plt.title("Number of Ratings of Highest Rated iPhones")
plt.xlabel("iPhone Model")
plt.ylabel("Number of Ratings")
plt.xticks(rotation=90)  # Optional: Rotate x-axis labels for better readability
plt.show()
```

According to the above bar graph, APPLE iPhone 8 Plus (Gold, 64 GB) has the most ratings

Now let’s have a look at the number of reviews of the highest-rated iPhones

```python
iphones = highest_rated["Product Name"].value_counts()
label = iphones.index
counts = highest_rated["Number Of Reviews"]

plt.figure(figsize=(6, 3))   
plt.bar(label, counts)
plt.title("Number Of Reviews of Highest Rated iPhones")
plt.xlabel("iPhone Model")
plt.ylabel("Number Of Reviews")
plt.xticks(rotation=90)  # Optional: Rotate x-axis labels for better readability
plt.show()
```

APPLE iPhone 8 Plus (Gold, 64 GB) is also leading in the highest number of reviews

Now let’s have a look at the relationship between the sale price of iPhones and their ratings 

```python

# Extract data from your DataFrame
sale_price = data["Sale Price"]
number_of_ratings = data["Number Of Ratings"]

# Create a scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(sale_price, number_of_ratings, alpha=0.5)  # Alpha sets marker transparency
plt.title("Relationship between Sale Price and Number of Ratings of iPhones")
plt.xlabel("Sale Price")
plt.ylabel("Number Of Ratings")
plt.grid(True)  # Optional: Add a grid
plt.show()
```

here is a negative linear relationship between the sale price of iPhones and the number of ratings. It means iPhones with lower sale prices are sold more

Now let’s have a look at the relationship between the discount percentage
on iPhones and the number of ratings:

```python
# Extract data from your DataFrame
Discount_Percentage = data["Discount Percentage"]
number_of_ratings = data["Number Of Ratings"]

# Create a scatter plot
plt.figure(figsize=(10, 6))
plt.scatter( number_of_ratings, Discount_Percentage, alpha=0.5)  # Alpha sets marker transparency
plt.title("Relationship between Sale Price and Number of Ratings of iPhones")
plt.xlabel("Number Of Ratings")
plt.ylabel("Discount Percentage")
plt.grid(True)  # Optional: Add a grid
plt.show()
```

There is a linear relationship between the discount percentage on iPhones and the number of ratings. It means iPhones with high discounts are sold more.
